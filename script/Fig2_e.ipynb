```python
import pandas as pd
from plot_utils import plot_violin
```

```python
df = pd.read_csv('../figure_data/Fig2_e.csv')
low_group = df[df['AI Score Classification']=='Low']
high_group = df[df['AI Score Classification']=='High']
```

```python
data = {
    'AI Score Low': low_group['Normlized Log2 (Validated Score)'],
    'AI Score High': high_group['Normlized Log2 (Validated Score)'],
}

data_long = []
for group, values in data.items():
    for value in values:
        data_long.append([group, value])
data_long = pd.DataFrame(data_long, columns=['Group', 'Value'])
```

```python
plot_violin(data_long)
```
